```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from matplotlib import rc

# rc('font',**{'family':'sans-serif','sans-serif':['Helvetica'], 'size':30})
rc('font',**{'family':'serif','serif':['Times']})
rc('text', usetex=True)


# Read the CSV file into a DataFrame
df = pd.read_csv('/home/sarith/Desktop/input_rdf.csv')

# Extract columns 1, 2, 3, and 4 from the DataFrame
x1 = df.iloc[:, 0]
y1 = df.iloc[:, 1]
x2 = df.iloc[:, 2]
y2 = df.iloc[:, 3]

# Plot column 1 and 2 as a thick line plot
plt.plot(x1, y1, linewidth=2, label='This Work')

# Plot column 3 and 4 as a scatter plot with red color and label
plt.scatter(x2, y2, color='red', label='C.J. Tainter et al.')

# Set axis labels and tick font
plt.xlabel(r'r ($\AA$)', fontsize=30)
plt.ylabel(r'g(r)$_{o-o}$', fontsize=30)

# Remove title
# plt.title('', fontname='Times New Roman', fontsize=14)

# Add legend
plt.legend()
# 
# Set figure size to A4
fig = plt.gcf()
fig.set_size_inches(11.69, 8.27) # A4 size in inches

# Add reference with clickable link
#plt.annotate('(1) C.J. Tainter et al., doi: 10.1063/1.3587053', xy=(0.5, 0.02), xycoords='axes fraction',
             #url='https://doi.org/10.1063/1.3587053', color='blue', fontname='Times New Roman', fontsize=10)

# Add AutoMinorLocator to the x-axis
plt.gca().xaxis.set_minor_locator(ticker.AutoMinorLocator())

# Set tick font
plt.tick_params(axis='both', which='major', labelsize=30, direction='in', top=True, right=True, bottom=True, left=True, width=1.5, length=5, pad=5)
plt.tick_params(axis='both', which='minor', direction='in', top=True, right=True, bottom=True, left=True, width=1, length=3)

# Set font for all text in the plot
# plt.rcParams['font.family'] = 'sans-serif'

# Save the plot as a PDF file
plt.savefig('plot.pdf', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```
